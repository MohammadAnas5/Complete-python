## Data Frame 
**2-d structure of data, consist of rows and columns (table in pandas)** 
######  one column or one row  is called series
`df['cost'].apply(fun)` 
- This code is written in Python and uses the Pandas library. Here's a step-by-step explanation of what it does:
1. df is a pandas DataFrame object, which contains a table of data with rows and columns.
2. The code accesses the 'cost' column of the DataFrame by using the indexing operator [].
3. The apply method is then called on the 'cost' column, which applies the given function fun to each element in the column.
4. The function fun takes each value in the 'cost' column as an argument and performs some operation on it. The exact operation performed by the function depends on how it's defined.
5. The result of the apply method is a new column with the same number of rows as the original 'cost' column, where each element is the result of applying the function fun to the corresponding element in the 'cost' column.
6. The transformed values in the new column replace the original values in the 'cost' column.`

### `apply()` on more then one column 

```python
data  = {
    'pointA': [(1,-2),(2,1),(1,3),(4,-3)],
    'pointB':[(3,-1),(4,-1),(-3,3),(2,-1)]
}

point_data = pd.DataFrame(data) 
point_data 
```

```python
def euclidian(row):
    a = row['pointA']
    b = row['pointB'] 
    return ((a[0]-b[0])**2 + (a[1]-b[1])**2)**0.5 
```

### `**axis=1** means take row by row `

```python
point_data.apply(euclidian,axis=1)  
```

```python
point_data['Euclidian_distance']=point_data.apply(euclidian,axis=1)   
```
