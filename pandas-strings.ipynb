```python
titanic = pd.read_csv('/kaggle/input/titanic123/ChanDarren_RaiTaran_Lab2a.csv')  
titanic 
```

```python
titanic['Name'] 
```

```python
# common functions 
# upper(), lower(), capitalize(), title() 
titanic['Name'].str.lower() 
titanic['Name'].str.upper() 
titanic['Name'].str.capitalize() 
titanic['Name'].str.title() 
```

```python
# len 
# -- largest name
titanic['Name'][titanic['Name'].str.len()==82].values[0]     
```

```python
# strip -- removing the leading and laging space 
"           anas          ".strip() 
```

```python
titanic['Name'].str.strip() 
```

```python
# split() and get() 
# fetch the last name, title(salutation), first name 
titanic['last_name']=titanic['Name'].str.split(',').str.get(0)  
```

```python
titanic[['title','first_name']] = titanic['Name'].str.split(',').str.get(1).str.strip().str.split(" ", n=1, expand=True) 
titanic 
```

```python
# replace function
titanic['title'] = titanic['title'].str.replace('Mlle.','Miss.')
titanic['title'] = titanic['title'].str.replace('Ms.','Miss.')  
```

```python
titanic['title'].value_counts() 
```

```python
# endswith & startswith 
titanic[titanic['first_name'].str.startswith('A')]  
```

```python
titanic[titanic['first_name'].str.endswith('A')]  
```

```python
titanic[titanic['first_name'].str.isnumeric()]   
```

```python
# applying regex 
# search john --> both case 
titanic[titanic['first_name'].str.contains('john', case=False)] 
```

```python
# lastname starts and with vowel 
titanic[titanic['last_name'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]  
```

```python
# for non vowel - consonant 
titanic[titanic['last_name'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')] 
```
