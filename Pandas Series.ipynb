### Pandas Series - object and datatype of pandas

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.(1-d array with index)

```python
import pandas as pd 
import numpy as np 
```

### Creating Series from List

```python
# string (or object 90% call with this name)
a = ["india","pakistan","USA","Australia","AIE"]
pd.Series(a) 
```

```python
# integer (int64 or int32)
a = [45,57,33,100]
pd.Series(a) 
```

```python
# custom index 
subject = ["English","Maths","Urdu","Science"]
marks = [56,55,94,80] 

pd.Series(marks,index=subject) 
```

```python
# give it a name 
pd.Series(marks,index=subject,name="Personal Marks") 
```

### Create series from dictionary

```python
a = {
    "maths":55,
    "english":66,
    "Science":77
}

a_series = pd.Series(a,name="new marks") 
a_series
```

## apply() 
apply method allow the users to pass a function and apply it on every single value of the Pandas series.

```python
movies.apply(lambda x: x.split()[0].upper()) 
```

```python
subs.apply(lambda x: "good days" if x>subs.mean() else "bad days" ) 
```
