```python
# GroupBy Attributes and Methods
# find total number of groups -> len
# find items in each group -> size
# first()/last() -> nth item
# get_group -> vs filtering
# groups
# describe
# sample
# nunique
```

```python
# find total number of groups -> len 
len(i1000.groupby('Genre'))
```

```python
i1000['Genre'].nunique() 
```

```python
#  find items in each group -> size
i1000.groupby('Genre').size()
```

```python
# first()/last() -> nth item 
i1000.groupby('Genre').first() 
i1000.groupby('Genre').last() 
i1000.groupby('Genre').nth(4)   
```

```python
# i1000[i1000['Genre']=='Biography']
i1000.groupby('Genre').get_group('Biography')  
```

```python
i1000.groupby('Genre').sample(1) # one random row in every group 
i1000.groupby('Genre').describe() # explain every numerical column group wize 
i1000.groupby('Genre').nunique()  # no. of unique value in every group 
```

```python
# agg function -- for different aggregate function on different column after group by 
i1000.groupby('Genre').agg({
    "Runtime":'mean',
    "IMDB_Rating": "mean",
    "No_of_Votes":"sum",
    "Gross":"sum"   
})
```

```python
# passing list 
i1000.groupby('Genre').agg(['min','max','sum']) 
```

```python
# combining the both 
i1000.groupby('Genre').agg({
    "Runtime":['mean','max'],
    "IMDB_Rating": ["mean","min"],
    "No_of_Votes":"sum",
    "Gross":["sum","mean"]    
})
```

# ` movies['Series_Title'].str.startswith('A') : Work on column work row wise` 
# ` apply() : work group by group (not row by row) - decorator of groups

```python
# split (apply) combine
# apply -> builtin function

i1000.groupby('Genre').apply(min) 
```

```python
def fun(group):
    return group['Series_Title'].str.startswith('A').sum()     
```

```python
# find number of movies starting with A for each group 

i1000.groupby('Genre').apply(fun)  
```

```python
t1 = i1000[i1000['Series_Title'].str.startswith("A")] 
t1.value_counts('Genre') 
```

```python
# find ranking of each movie in the group according to IMDB score 
def fun(group):
    group['genre_ranking']=group['IMDB_Rating'].rank(ascending=False) 
    return group 
```

```python
i1000.groupby('Genre').apply(fun) 
```

```python
def normal(group):
    group['Normalized_IMDB'] = (group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
    
    return group 
```

```python
# find normalizd IMDB rating of every group (genre)
i1000.groupby('Genre').apply(normal)
```

```python
# Group by more then one column 
i1000.groupby(['Director','Star1']).size()  
i1000.groupby(['Director','Star1']).get_group(('Aamir Khan','Amole Gupte') ) 
```

```python
# find the most earning actor->director combo 
i1000.groupby(['Star1','Director'])['Gross'].sum().sort_values(ascending=False) 
```

```python
i1000['Metascore'] = i1000['Metascore'].fillna(i1000['Metascore'].mean())   
```

```python
# find the best(in-terms of metascore(avg)) actor->genre combo 
i1000.groupby(['Star1','Genre'])['Metascore'].mean().reset_index()   
```

```python
# with group by 
i1000.groupby(['Star1','Genre']).agg(['mean','min','max']) 
```

```python
# Create a function that can return the highest score of any batsman 
df1 = ipl[ipl['batsman']=='V Kohli'] 
df1.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]  
```

```python
def highest(player):
    df1 = ipl[ipl['batsman']==player] 
    return df1.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]     
```
