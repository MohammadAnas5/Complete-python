# Scatter Plot
It is mainstay of statistical visulization, It depict the joint distribution of 2 variable or 3 variable where each point represent the information of 1 row or an observation, By this plot we easily see whether there is any relationship between them or not. 

# 

It can be plotted for numerical and categorical data but we generally use it for numerical data
.There are two ways to create scatter plot (seaborn have 2 function for creating scatter plot) --
1. replot (relational plot) [Figure Level function] - have both scatter and line plot
2. scatterplot [Axes Level function]

it have 3 additional variable or 3 turning knob of points or marker names **hue,size and style**.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
%matplotlib inline 
```

```python
plt.style.use("fivethirtyeight")
```

```python
data = sns.load_dataset("tips") 
```

```python
data.head() 
```

## Plot a scatter plot b/w total_bill and tip

```python
sns.relplot(x="total_bill",y='tip',kind="scatter", data=data) 
```

```python
# using the scatter plot function
sns.scatterplot(x="total_bill",y='tip', data=data) 
```

## With hue parameter 
Hue can be used to group to multiple data variable and show the dependency of the passed data values are to be plotted.
(it add more categoryor information to plot) 
- Generally we pass categorical data

```python
sns.relplot(x="total_bill", y="tip", data=data, hue="smoker") 
```

## With style parameter 
Add more piece of information to plot
- Generally we pass categorical data

```python
sns.relplot(x="total_bill", y="tip", data=data, hue="smoker", style="sex")  
```

## With size parameter 
Add more piece of information by change in size of points to plot
- Generally we pass continous numerical data

```python
sns.relplot(x="total_bill", y="tip", data=data, hue="smoker", style="sex", size="size") 
```

```python
# change the size of marker
sns.relplot(x="total_bill", y="tip", data=data, hue="smoker", style="sex", size="size", sizes=(15,200))  
```

```python
#1. Strip plot -- A strip plot is a graphical data anlysis technique for summarizing a univariate data set
sns.catplot(x="day", y="tip", data=data, kind="strip", jitter=1, palette="spring")     
# jitter - to avoid overlapping and points are easily visible, give some width or strip to x axis feature by adding noise
```

```python
sns.stripplot(x="day", y="tip", data=data) 
```

```python
# 2. swarm plot -- A swarm plot is very similar to a strip plot, yet the locations of points are adjusted automatically 
#                                      to avoid overlap even if the jitter value is not applied.
# When multiple values of x column occupying the same value of y columns then it arrange all values in such a manner 
# so it show distribution also

sns.catplot(x='day',y="tip", data=data, kind="swarm") 
```

```python
sns.swarmplot(x="day",y="tip",data=data) 
```

```python
# use hue parameter
sns.catplot(x='day',y="tip", data=data, kind="swarm", palette="magma", hue="sex")     
```

## 1. box plot -- numerical and categorical
Boxplots are a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”).


PDF's is used to check the centrality of data

```python
# Univariate analysis
# sns.catplot('total_bill',data=data, kind='box')
sns.boxplot(data['total_bill']) 
```

```python
# Bivariate -- see the distribution in b/w 2 column 
# sns.catplot('day','total_bill',data=data, kind='box')
sns.boxplot(data['day'],data['total_bill'])
```

## 2. Violin plot -- distribution of data 
  A violin plot shows distributions of numeric data (for one or more groups using density curves) and it is a combination of box plot and probability density functions (PDF's). 
  
  The width of each curve corresponds with the approximate frequency of data points in each region. Densities are frequently accompanied by an overlaid chart type, such as box plot, to provide additional information.
  
  

### Its often used to compare the distribution of variable in different groups

```python
sns.violinplot(data['tip']) 
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=data)
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=data, hue='sex') 
```

```python
# both left and right side are equal in violin -- we also take only only one side
sns.catplot(x='day', y='total_bill', kind='violin', data=data, hue='sex',split=True)  
```
