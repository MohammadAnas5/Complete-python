## 3D Surface Plot 

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100) 
xx,yy = np.meshgrid(x,y) #2d numpy array
xx.shape  
```

```python
z = xx**2+yy**2 
```

```python
fig = plt.figure()
ax = plt.subplot(projection='3d') 
p = ax.plot_surface(xx,yy,z,cmap='viridis') 
fig.colorbar(p) 
```

```python
z= np.sin(xx)+np.cos(yy) 
fig = plt.figure()
ax = plt.subplot(projection='3d') 
p = ax.plot_surface(xx,yy,z,cmap='viridis') 
fig.colorbar(p) 
```

## Contour Plot 
plot 3d in 2d
